```python
#Type of the string using the type function
type(multiline_text)
```

```python
# Length of the string 
len(multiline_text)
```

```python
#Remove all new lines
multiline_text = multiline_text.replace('\n',"")
multiline_text
```

```python
# Remove special characters and punctuation
clean_multiline_text =""
for char in multiline_text:
    if char=="":
        clean_multiline_text+=char
    elif char.isalnum():
        clean_multiline_text+=char
    else:
        clean_multiline_text+=" "
            
```

```python
# Contents of the cleaned multiline text is displayed below
clean_multiline_text
```

```python
# Generate a list of all the words
list_words = clean_multiline_text.split()
list_words
```

```python
# Number of words in the list
len(list_words)
```

```python
# Create a list from the list created above
unique_words_dict = dict.fromkeys(list_words)
len(list(unique_words_dict.keys()))
```

```python
# Count the number of times each unique words appeared in the cleaned text
for word in list_words:
    if unique_words_dict[word] is None:
        unique_words_dict[word] = 1
    else:
        unique_words_dict[word]+=1
unique_words_dict
        
```

```python
# Find the top 25 words from the unique word
top_words = sorted(unique_words_dict.items(), key=lambda key_val_tuple:key_val_tuple[1],reverse=True)
top_words[:25]
```

```python
# 5. Check the actual type that dropwhile returns
import math
def convert_to_a_number(number_stack):
    final_number = 0
    for i in range(0, len(number_stack)):
        final_number+=(number_stack.pop()*(math.pow(10,i)))
    return final_number
for number_tuple in permutations(range(3)):
    number_stack = list(dropwhile(lambda x:x<=0,number_tuple))
    print(convert_to_a_number(number_stack))
```

```python
#1 Import "zip_longest" from itertools
from itertools import zip_longest
```

```python
# 1b. Write a function that contains header, line and fillvalue as None
def return_dict_from_the_csv_line(header, line):
    zipped_line = zip_longest(header, line, fillvalue=None)
    ret_dict = {kv[0]:kv[1]for kv in zipped_line}
    return ret_dict    
```
